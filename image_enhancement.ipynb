```python
from PIL import Image, ImageEnhance

def adjust_brightness_contrast(image_path, brightness_factor, contrast_factor):

  img = Image.open('download (3).jpeg')

  enhancer = ImageEnhance.Brightness(img)
  img = enhancer.enhance(brightness_factor)

  enhancer = ImageEnhance.Contrast(img)
  img = enhancer.enhance(contrast_factor)

  return img

brightness_factor = 0.5
contrast_factor = 1.2

adjusted_image = adjust_brightness_contrast(img, brightness_factor, contrast_factor)

adjusted_image
```

```python
import numpy as np
import cv2

img = cv2.imread('download (3).jpeg')

blue, green, red = cv2.split(img)

red_adjusted = cv2.addWeighted(red, 1.5, 0,0,0)

adjusted_img = cv2.merge((blue, green, red_adjusted))

adjusted_img
```

```python
import cv2
import numpy as np

img = cv2.imread('download (3).jpeg')

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

hsv[:, :, 0] = (hsv[:, :, 0] + 50) % 180

hue_shifted_img = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
hue_shifted_img
```
